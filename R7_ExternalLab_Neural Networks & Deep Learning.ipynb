```python
import h5py    
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import keras
from keras.models import Sequential
from keras.layers import Dense, Conv2D,Conv3D, MaxPooling2D,MaxPooling3D, Dropout, Flatten, Activation
from keras.layers.normalization import BatchNormalization
from keras.preprocessing import image
import tensorflow as tf
```

```python
df = h5py.File("SVHN_single_grey1.h5",'r') 
```

```python
print(list(df.keys()))
```

```python
shape = df["y_train"].shape
shape
```

```python
X_test = df['X_test'][:]
y_test = df['y_test'][:]
X_train = df['X_train'][:]
y_train = df['y_train'][:]
y_val = df['y_val'][:]
X_val = df['X_val'][:]
```

```python
df.close()
```

```python
print('Training set', X_train.shape, y_train.shape)
print('Validation set', X_val.shape, y_val.shape)
print('Test set', X_test.shape, y_test.shape)
```

```python
y_train = keras.utils.to_categorical(y_train, 10)
y_test = keras.utils.to_categorical(y_test, 10)
```

```python
X_train =X_train.reshape(X_train.shape[0],32,32,1).astype('float32')
X_test =X_test.reshape(X_test.shape[0],32,32,1).astype('float32')
```

```python
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
X_train /= 255
X_test /= 255
```

```python
print('X_train shape:', X_train.shape)
print('No: of images in Train Dataset = ', X_train.shape[0])
print('NO: of images in Test Dataset = ', X_test.shape[0])
```

```python
model=Sequential()
model.add(Conv2D(32,(3,3),border_mode='valid',input_shape=(32,32,1),activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.2))
```

```python
model.add(Flatten())
model.add(Dense(128,activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(10,activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy',optimizer='sgd',metrics=['accuracy'])
```

```python
model.fit(X_train,y_train,batch_size=128,nb_epoch=10,verbose=1,validation_data=(X_test,y_test))
```

```python
score=model.evaluate(X_test,y_test)
score
```

```python
predict = model.predict(X_test)
```
