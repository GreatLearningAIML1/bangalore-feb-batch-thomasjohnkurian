### Build a DNN using Keras with `RELU` and `ADAM`

#### Load tensorflow

```python
import tensorflow as tf
```

#### Collect Fashion mnist data from tf.keras.datasets 

```python
from keras.datasets import fashion_mnist
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()
```

#### Change train and test labels into one-hot vectors

```python
trainY = tf.keras.utils.to_categorical(y_train, num_classes=10)
testY = tf.keras.utils.to_categorical(y_test, num_classes=10)
```

#### Build the Graph

#### Initialize model, reshape & normalize data

```python
x_train = x_train.astype('float32') / 255
x_test = x_test.astype('float32') / 255
```

```python
x_train=x_train.reshape(x_train.shape[0],28,28,1).astype('float32')
x_test=x_test.reshape(x_test.shape[0],28,28,1).astype('float32')
```

```python
tf.keras.backend.clear_session()
model = tf.keras.models.Sequential()
```

```python
model.add(tf.keras.layers.Conv2D(32, kernel_size=(3,3),  activation='relu'))
model.add(tf.keras.layers.Flatten())
```

#### Add two fully connected layers with 200 and 100 neurons respectively with `relu` activations. Add a dropout layer with `p=0.25`

```python
model.add(tf.keras.layers.Dropout(0.25))
model.add(tf.keras.layers.Dense(200, activation='relu'))
model.add(tf.keras.layers.Dense(100, activation='relu'))
```

### Add the output layer with a fully connected layer with 10 neurons with `softmax` activation. Use `categorical_crossentropy` loss and `adam` optimizer and train the network. And, report the final validation.

```python
model.add(tf.keras.layers.Dense(10, activation='softmax'))
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x_train,trainY,          
          validation_data=(x_test,testY),
          epochs=10,
          batch_size=32)
```

```python
# Evaluate the model on train set
score2 = model.evaluate(x_train, trainY, verbose=0)
# Print test accuracy
print('\n', 'Train accuracy:', score2[1])
```

```python
# Evaluate the model on test set
score = model.evaluate(x_test, testY, verbose=0)

# Print test accuracy
print('\n', 'Test accuracy:', score[1])
```

```python
During the start of first epoc the loss was .38 which went on to improve till .05 by the end of 10th epoch. Also the accuracy level went above 90% after all the training.
```
