# Use the five percent significance test over the data to determine the p value to check new scheme has significantly raised outputs? (10 points)

```python
from scipy.stats import ttest_rel, ttest_ind, levene, shapiro
from statsmodels.stats.power import ttest_power
```

```python
# Seperating data into 2 groups
old_scheme = data['Old Scheme (in thousands)'].values.reshape(-1,1)
new_scheme = data['New Scheme (in thousands)'].values.reshape(-1,1)
```

```python
#For checking equality of variance between old_scheme and new_scheme
levene(old_scheme,new_scheme)
```

#Since the pvalue of levene test is greater than .05 the null hypothesis Variences are equal stands true.

```python
# For checking Normality distribution of old_scheme
shapiro(old_scheme)
```

#Since pvalue of shapiro test is greater than .05 null hypothesis "Distribution is Normal" stands true for old_scheme

```python
# For checking Normality distribution of new_scheme
shapiro(new_scheme)
```

#Since pvalue of shapiro test is greater than .05 null hypothesis "Distribution is Normal" stands true for new_scheme

```python
#We have to test differences between samples of the same group at different points in time so we can use the Paired T Test,since old_scheme and new_scheme has same variance 
#null hypothesis: old_scheme and new_scheme has the same mean
t_statistic, p_value = ttest_rel(old_scheme,new_scheme)
print(t_statistic, p_value)
```

Since the p value is greater than .05 the null hypothesis "old_scheme and new_scheme has the same mean" stands true and we can assume that there is no significance difference between the mean of old scheme and new scheme. so Titan won't get the break even with the current performance of their sales persons.

# The probability of a type 1 error? (2.5 points)

The type I error rate or significance level is the probability of rejecting the null hypothesis given that it is true.
In our case the significance level is set to 0.05 (5%), implying that it is acceptable to have a 5% probability of incorrectly rejecting the null hypothesis.
So the probability of a type 1 error is 5%.
